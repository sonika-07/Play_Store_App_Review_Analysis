```python
df.shape
```

c. **Rating**

Null values of rating are to be filled with 0 as it seems that users are not interested in that app

```python
#Fill the null values with 0
df['Rating']=df['Rating'].fillna(0)
```

d. **Reviews**

```python
df['Reviews'].unique()
```

```python
#Change the datatype of reviews.
df['Reviews']=df['Reviews'].astype(float)
```

 **e. Installs**

```python
#Checking for unique values in installs column
df.Installs.unique()
```

```python
#Removing the special characters '+' , ',' and changing the datatype

char_to_remove=['+',',']
for char in char_to_remove:
  df['Installs']=df['Installs'].apply(lambda x:x.replace(char,""))
df['Installs']=df['Installs'].astype(int)
```

*Check if there are fake reviews*

```python
df[df.Reviews>df.Installs]
```

```python
df[df.Reviews>df.Installs].shape
```

*Maximum possible no. of reviews cannot exceed the total no. of installs of particular app, therefore replacing it with no. of installs*

```python
#Replacing those values of reviews which are greater than intalls with Number of installs
df.loc[df['Reviews']>df['Installs'],'Reviews']=df['Installs']
```

```python
df[df.Reviews>df.Installs].shape
```

*Now there is no fake review in the table*

f. **Size**

```python
#Unique values of size
df['Size'].unique()
```

```python
#Removing the secial characters '+' , ',' , 'k' , 'M' , 'Varies with device'

df['Size']=df.Size.apply(lambda x:x.strip('+'))
df['Size']=df['Size'].apply(lambda x:x.replace(',','')) 
df['Size']=df.Size.apply(lambda x: x.replace('k','')) 
df['Size']=df['Size'].apply(lambda x:x.replace('M','000')) 
df['Size']=df.Size.replace('Varies with device',np.NaN) 
```

```python
#Changing the datatype of size column

df['Size'].fillna(str(df['Size'].mode()[0]),inplace=True)
df['Size']=df['Size'].astype(float)
```

g. **Type**

```python
#unique values of type of apps

df.Type.unique()
```

```python
#see if there are null values in type column

df[df['Type'].isna()]
```

*Since the Price of Row number 9148 is 0.0 ,hence we can fill the type with 'Free'.*

```python
#fill the null value with free
df['Type']=df['Type'].fillna('Free')
```

h. **Price**

*Remove special characters from Price*

```python
df.Price.unique()
```

```python
#removing the special characters from price ',' , '$' and changing the datatype

char_to_remove=[',','$']
for char in char_to_remove:
  df['Price']=df['Price'].apply(lambda x:x.replace(char,""))
df['Price']=df['Price'].astype(float)
```

```python
df.isnull().sum()
```

i. **Handling current and android version column**

```python
df['Current Ver'].value_counts()
```

```python
df.loc[df['Current Ver']=='Varies with device','Current Ver']=np.nan
```

```python
#Fill null values of current version with mode

df['Current Ver'].fillna(str(df['Current Ver'].mode()[0]),inplace=True)
```

```python
df['Android Ver'].value_counts()
```

```python
df.loc[df['Android Ver']=='Varies with device','Android Ver']=np.nan
```

```python
#Fill null values of android version with mode

df['Android Ver'].fillna(df['Android Ver'].mode()[0],inplace=True)
```
